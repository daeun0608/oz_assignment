## 🤍 DB
- DB=데이터 저장소=데이터베이스, 데이터의 집합
- 효율적인 데이터 관리
- 여러 사용자가 실시간으로 데이터 사용이 가능
- DB라는 것이 있기 전에는 파일 형태로 데이터를 관리
    - 파일 형태의 가장 큰 문제점은 데이터를 업데이트 하면 기존 버전의 데이터를 알 수 없다는 것.
<br><br>
#### **데이터 모델** : 데이터를 어떻게 저장하고 관리할지에 대한 방법의 차이
1. 계층형 데이터 모델 : 윈도우 파일 시스템의 디렉터리 구조
2. 네트워크형 데이터 모델 : 파일을 옮길 때 루트까지 가야하는 계층형 데이터 모델을 보완한 모델<br>
여러 부모 레코드를 가질 수 있지만 부모와 자식간의 관계가 복잡한 구조
3. **관계형 데이터 모델** : 가장 널리 사용되는 데이터 모델
    - 테이블 형태 ( 레코드(행, row)과 필드(열, column)로 구성 )
    - SQL(Structured Query Language)을 사용하여 데이터를 조작
    - 관계형 데이터베이스 시스템(RDBMS)에서 지원
4. 객체 지향형 데이터 모델 : 데이터와 해당 동작을 함께 캡슐화하며, 객체간의 관계를 강조

## 🤍 DBMS
데이터베이스 관리 시스템(Database Management System)
- 데이터의 저장, 검색, 업데이트, 관리를 위한 소프트웨어
- 대용량 데이터의 효율적인 처리와 접근을 가능하게 하며, 데이터의 무결성, 보안, 백업 및 복구 기능을 제공
<br><br>
#### **주요 DBMS 종류 및 특징**
1. Oracle Database
    - MySQL을 인수
    - 비싼 라이선스 비용이 있지만, 대규모 엔터프라이즈 환경에서의 신뢰성과 성능은 매우 높아 대부분의 산업 분야에서 사용
2. MySQL
    - 오픈 소스이며, 가장 인기 있는 데이터베이스 중 하나
    - 소규모에서 중간 규모의 프로젝트에 사용 ( 대규모 엔터프라이즈 시스템에 이르기까지 다양한 용도로 사용 )
    - LEMP : 오픈소스로 백엔드를 구축해서 Linux, Nginx, MySQL, PHP를 확용해 만든 소프트웨어 스택 서비스
3. Microsoft SQL Server
    - .NET과 같은 Microsoft 기술과 잘 통합
    - Windows 기반 시스템에 최적화되어 있어, Microsoft 환경에서 뛰어난 성능을 발휘
    - 데이터 웨어하우징, 비즈니스 인텔리전스, 데이터 분석에 유용한 기능을 포함
        - 직접 해야하는 것을 유료로 이용할 수 있음
4. PostgreSQL - 고급 데이터베이스 요구에 적합
5. MongoDB
    - 대용량 데이터 처리와 실시간 분석에 최적화 ( 빅데이터를 겨냥함 )
    - NoSQL 데이터베이스 중 하나로, 문서 지향적 구조 ( JSON 형식의 문서 )
    - 스키마가 없어 데이터 구조가 유연하고, 개발과 확장이 쉬움
6. Redis (Remote Dictionary Server)
    - 키-값 저장소
    - 캐싱 서버의 이용으로 다수의 사용자와 웹소켓으로 연결하지 않고 일정시간마다 푸시 / 풀하면서 실시간연결에 사용된다.
    - 인메모리 데이터 스토어로, 매우 빠른 읽기와 쓰기 속도를 제공 ( 디스크에 데이터를 저장해서 유실 가능성 낮음 )
7. Elasticsearch
    - 로그 및 데이터 집계, 분석에 사용
    - ELK(Elasticsearch, Logstash, Kibana) 스택으로 사용
8. SQLite
    - 경량이면서도 서버 없이 로컬에서 사용할 수 있는 RDBMS
    - 임베디드 시스템이나 모바일 애플리케이션에서 많이 활용
9. Snowflake
    - 대규모의 데이터 저장 특화
    - 소프트웨어 설치 없이 사용
10. Cassandra
    - 대규모 분산 데이터베이스( 데이터는 여러 노드에 걸쳐 분산 저장 )로, 큰 규모의 데이터를 관리
11. MariaDB
    - MySQL의 포크인 오픈소스로 커뮤니티 중심의 개발 모델
12. Apache Hive
    - Hadoop( 데이터의 양이 많음, DB의 부담을 줄임 ) 위에 구축된 데이터 웨어하우스 시스템
13. Google BigQuery
    - BigQuery ML을 통해 SQL 쿼리 내에서 머신러닝 모델을 생성하고 실행 가능

## 🤍 RDBMS(Relational DBMS)
- 관계형 데이터베이스 관리 시스템 : 테이블 형태로 데이터를 구조화하고 관리하는 데이터베이스
- 단점
    - 성능 : 대용량 대이터 처리나 복잡한 쿼리 성능이 다른 모델에 비해 비교적 낮음
    - 확장 어려움 : 수직은 한계, 수평은 복잡하고 비용이 많이 듦
    - 유연성 부족 : 스키마 변경이 어려워서 요구사항 변경에 대응 어려움
    - 복잡한 조인 : 많은 테이블 간의 조인은 성능에 부정적 영향
    - 고비용 : 서버 하드웨어 및 라이선스 비용이 상대적으로 높음
- 대표적인 RDBMS : MySQL, PostgreSQL, Microsoft SQL Server, Oracle Database, SQLite
<br><br>
### RDBMS 기본 구조(Structure)
1. **스키마(Schema)**<br>데이터베이스의 구조를 정의<br>테이블, 행, 열, 인덱스, 관계 등의 데이터베이스 구조를 명시
2. **SQL(Structured Query Language)**<br>데이터베이스에서 데이터를 관리하기 위해 사용되는 표준 프로그래밍 언어
3. **인덱스(Index)**<br>데이터베이스에서 데이터 검색 속도를 빠르게 하기 위해 사용되는 객체<br>인덱스는 특정 열에 대한 포인터를 포함하여 데이터 검색 시간을 단축
4. **데이터베이스 관리 시스템(DBMS)**<br>데이터베이스를 생성하고 관리하는 소프트웨어입니다<br>ex) MySQL, PostgreSQL, Oracle, Microsoft SQL Server
5. **트랜잭션(Transaction)**<br>데이터베이스의 상태를 변화시키는 하나의 작업 단위
```
트랜잭션(하나의 작업 단위) 특징 (ACID) : 데이터베이스의 트랜잭션이 안전하게 수행되는 것을 보장하는 성질을 가리킴
1. 독립성 (원자성Atomicity)
    - 물리적, 논리적 독립성
    - 트랜잭션이 하나의 원자처럼 동작함, 연산이 모두 수행되거나 모두 수행되지 않음 all or nothing
2. 일관성Consistency
    - 데이터의 불일치성을 배체 (트랜잭션 전후로 DB가 일관되어 데이터가 조건에 위배되지 않음)
    - 데이터의 정확성과 일관성을 보장 **데이터 무결성**을 구현
3. 고립성Isolation
    - 트랜잭션 작업 수행은 고립된 상태에서 수행되어 다른 트랜잭션에 영향을 주어서도, 간섭을 받아서도 안 된다는 것
4. 지속성Durablility
    - 트랜잭션이 완료되는 시점에 조작이 영구적이 되어 결과를 잃지 않는 것
    - 시스템이 정상일 때 뿐 아니라 데이터베이스나 OS의 이상 종료, 즉 시스템 장애도 견딜 수 있다는 것
```
<br><br>
### RDBMS 구성 요소
- Database → Table → Column, Row
1. Database : 구조화된 데이터의 집합
2. Table : 데이터베이스 내에서 데이터가 저장되는 장소
3. 테이블은 열(Column)과 행(Row)으로 이뤄져 있으며,<br>각 열은 데이터의 속성을 나타내고 각 행은 해당 속성에 대응하는 데이터 값을 포함<br>Column = Attribute = Field (열)<br>Row = Record = Tuple (행)
    - 기본 키(Primary Key) : 테이블 내의 각 행을 고유하게 식별하는 열
    - 외래 키(Foreign Key) : 다른 테이블의 기본 키를 참조하는 열, 두 테이블 간의 관계를 설정하는 데 사용

## 🤍 데이터베이스 스키마Schema
데이터베이스에서 데이터의 구조와 구성을 정의하는 청사진 또는 설계도<br>
일반적으로 관계형 데이터베이스에서는 여러 테이블로 구성되며, 각 테이블의 속성을 식별하여 컬럼으로 정의
<br><br>
**논리적 스키마** : 테이블 간의 관계, 각 테이블의 속성과 데이터 타입, 제약 조건 등<br>
물리적 스키마 : 데이터의 실제 저장과 관련이 있는 부분으로, 디스크에 저장하는 방식 등
