## ♥︎ NoSQL (Not Only SQL)

#### NoSQL의 등장
1. **다양한 형태의 데이터가 중요해짐**<br>
관계형 DB는 테이블 형태로 정형화된 데이터에 적합, 현대 애플리케이션에서는 `반정형/비정형 데이터의 중요성`이 올라감.
문서, 그래프, 키-값 등 다양한 형태의 데이터를 다루기에 NoSQL이 더 유연하게 대처 가능

2. **데이터 저장 단위가 기하급수적으로 증가하면서 데이터 분산 처리가 중요해짐**<br>
대규모의 데이터를 효과적으로 처리하고 확장하기 위한 목적에서 등장<br>
NoSQL 데이터베이스는 수평적 확장이 용이, `분산 DB로서 대규모 트래픽과 데이터 다루는데 특히 유용`
> 즉, NoSQL은 **RDBMS의 한계**를 극복하고 텍스트 이외의 다양한 형태의 데이터를 처리하기 위해 등장
>
> **RDBMS의 단점**
> 1. 규모 확장성: 빅데이터와 분산 시스템에 대한 지원이 제한적<br>수직 확장(서버 성능 향상)은 좋지만, 수평 확장(서버 수 증가)에는 비효율<br>⇒ 클라우드 서버의 등장으로 대부분 해결
> 2. `유연성 부족`: 데이터 스키마가 고정되어 있어, 변경이나 새로운 데이터 타입의 추가가 어려움.<br>비구조화된 데이터나 변화하는 데이터 구조를 다루는 데 불편함을 초래
> 3. 복잡한 쿼리와 트랜잭션: (특히 대규모 분산 환경에서)복잡한 쿼리와 트랜잭션이 시스템의 성능에 영향을 미침<br>⇒ 데이터가 많은 경우 전체 데이터 조회 쿼리 한번으로 DB가 뻗어버릴 수 있음.


<br>

#### NoSQL의 특징
1. **스키마 유연성 (Schema Flexibility)**<br>
NoSQL 데이터베이스는 스키마가 동적(dynamic)이거나 덜 제한적이어서 데이터 모델을 쉽게 조정하고 변경 가능<br>
➡ 시스템의 확장 및 변경에 대처하기 좋음
➡️유연한 구조는 데이터를 여러 노드에 분산/저장해도 조회 및 저장에 큰 이슈를 없게함.

2. **분산 처리 (Distributed Processing)**<br>
NoSQL은 대부분 분산 데이터베이스로 설계 = 대량의 데이터를 여러 서버(노드)에 분산하여 저장/처리하도록 설계<br>
➡️단일 노드의 과부하 문제를 줄이고, 데이터의 병렬 처리를 향상

3. **높은 가용성과 확장성 (High Availability and Scalability)**<br>
NoSQL은 높은 가용성과 확장성을 제공하여 시스템이 확장될 때 성능의 감소가 없음

4. **다양한 데이터 모델**<br>
NoSQL은 다양한 데이터 모델을 지원(주요 모델 - 문서 지향, 키-값, 와이드 컬럼, 그래프)<br>
    - 키-값 저장 방식(많은 NoSQL 데이터베이스가 사용)<br>
      간단하고 빠른 데이터 액세스를 제공하여, 대규모 분산 시스템에서의 성능 향상에 기여

## ♥︎ NoSQL 데이터 타입
1. 정형 데이터(Structured Data)<br>정해진 스키마에 따라 구조화된 데이터 (RDB나 스프레드시트, CSV 같은 테이블 형식에서 사용)
<br><br>
2. 반정형 데이터(Semi-Structured Data)<br>일정한 구조는 있지만 엄격하게 준수하지 않는 데이터 (XML, JSON, HTML, 로그같은 형식에서 사용)
<br><br>
3. 비정형 데이터(Unstructured Data)<br>구조나 형태가 없으며, 연산이 어려운 데이터로, 영상, 이미지, 음성, 텍스트<br>
파일 경로를 지정하여 NoSQL에 저장하면 저장이 되지만, 보통은 S3(버킷데이터, 구글드라이브와 비슷함)와 같은 저장소에 업로드한 다음 링크를 저장하는 방식을 많이 사용

## ♥︎ NoSQL 종류
- **문서 지향 데이터베이스(Document Stores): MongoDB, CouchDB**
    - NoSQL에서 가장 많이 사용되는 타입 중 하나
    - 데이터를 JSON이나 BSON과 같은 문서 형식으로 저장하는 방식
    - 여러 필드를 포함하고 있어 복잡한 구조의 데이터를 표현 가능
        ```json
        {
          "_id": ObjectId("5f3c92c8e948af08996285a4"),
          "address": {
            "city": "New York",
            "zip": "10001"
          },
          "interests": ["reading", "traveling"]
        }
        ```
<br>

- **키-값 데이터베이스(Key-Value Stores): Redis, DynamoDB**
    - 키와 값의 쌍으로 데이터를 저장하는 방식
    - 주로 캐시나 설정 데이터와 같이 단순한 데이터에 사용
        ```perl
        "user:1234" => {"username": "alice", "email": "alice@example.com"}
        ```
<br>

- **와이드 컬럼 데이터베이스(Wide-Column Stores): Cassandra, HBase**
    - 열 기반의 데이터 모델(Column Family-테이블과 비슷한 역할)을 열의 집합으로 표현.
    - Bigtable과 같은 모델로, 여러 개의 컬럼 패밀리(Column Family)를 가진 테이블로 데이터를 저장<br><br>
        ```
      각 행은 다양한 수의 컬럼을 가질 수 있음
      슈퍼 컬럼 페밀리의 각 행은 하나 이상의 슈퍼 컬럼을 포함. (Row Key)
        | Row Key      | Column Family A        | Column Family B           |
        |--------------|------------------------|---------------------------|
        | user0004     | {"name": "Alice"}      | NULL                      |
        |--------------|------------------------|---------------------------|
        | user0005     | {"name": "Alice"}      | {"email": "alice@ex.com"} |
        ```
      <br><br>
        ```mathematica
      Column Family는 name의 집합이 아니라 비슷한 성격의 컬럼들의 그룹으로 묶은 것
      컬럼들을 주제별로 그룹화해서 Column Family로 나누는 게 와이드 컬럼 DB의 핵심
        | Row Key      | Column Family A            | Column Family B           |
        |--------------|----------------------------|---------------------------|
        | user1234     | { name: "Alice", age: 30 } | {"email": "alice@ex.com"} |
        |--------------|----------------------------|---------------------------|
        | user0004     | {"name": "Alice"}          | NULL                      |
        ```
<br>

- **그래프 데이터베이스(Graph Databases): Neo4j, Amazon Neptune**
    - 계형 데이터(부모-자식 관계)를 표현하기 위해 그래프 구조를 사용하는 데이터 모델
    - 노드(Node)와 엣지(Edge)로 구성(노드를 기반으로 데이터 연결하고 조합)

<br>

- 오브젝트 데이터베이스(Object-Oriented Databases): db4o, ObjectDB
    - 객체 지향 프로그래밍의 객체를 직접적으로 데이터베이스에 저장하고 관리 (저장과 사용이 쉽지 않아 많이 사용X)
    - 관계형 데이터베이스와는 다르게 객체 간의 관계, 상속, 다형성과 같은 객체 지향 프로그래밍의 개념을 그대로 지원
    - 객체를 저장하고, 객체 지향적인 쿼리를 사용하여 데이터를 검색이 가능해서, 객체를 더 직관적으로 표현

### NoSQL 사용 예시
**SQL과 비슷한 쿼리 언어를 제공하는 예시 프로그램**
1. Cassandra: Cassandra Query Language (CQL)을 제공, SQL과 유사한 구문
2. MongoDB: MongoDB는 전통적인 SQL 쿼리를 지원하지 않지만, 유사한 쿼리 구조를 가진 Aggregation Framework를 제공
<br><br>

**웹 어플리케이션의 세션 관리**<br>
1. **Redis** : 메모리 기반의 키-값 저장소로, 세션 데이터(사용자 ID, 로그인 시간 등)를 빠르게 저장하고 검색<br>
2. **flask, django등의 웹 프레임워크**들은 세션 기능을 제공하지만, `메모리가 아니라 서버에 데이터를 저장`<br>
    - ️개발이나 트래픽이 작은 어플에서는 충분하지만 사용자가 늘경우 느려지는 현상 발생<br>
    - 서버가 재시작되면 모두 삭제되지만 Redis는 디스크에 저장해서 데이터가 유실되지 않게 함️
<br><br>

**대규모 로그 및 이벤트 데이터 처리**<br>
**MongoDB+Elasticsearch**<br>
MongoDB : JSON 형식의 문서를 저장하는데 용이하며, 로그 데이터를 구조화하여 쉽게 쿼리하고 분석<br>
Elasticsearch : 로그 및 이벤트 데이터를 신속하게 색인화하여 검색 및 분석(데이터를 Kibana에서 시각화하여 사용)
<br><br>

**소셜 미디어 애플리케이션**<br>
**Neo4j와 같은 그래프 데이터베이스**는 친구 관계를 그래프 형태로 모델링하는 데 적합<br>
예를 들어, 소셜 미디어에서 사용자 간의 친구 관계를 노드와 엣지로 표현하여 친구 추천이나 소셜 그래프 분석에 활용
<br><br>

**클라우드 기반 애플리케이션**<br>
**AWS DynamoDB** : 클라우드 기반의 웹 어플리케이션에서 사용자 데이터를 효과적으로 저장하고 관리하는 데 활용<br>
(서버리스 아키텍처와의 통합을 통해 확장성과 가용성을 확보)
