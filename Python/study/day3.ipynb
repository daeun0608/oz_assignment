### 튜플
- 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 추가, 삭제, 변경 할 수 없다
    - 값을 추가, 삭제, 변경 할 수 없기 때문에 사용하지 못하는 메소드가 많다
- tuple은  tuple = 값, 값 / tuple = (값, 값)    으로 생성한다
- tuple은 소괄호이다
- len, index, count, 슬라이싱이 가능함
  
### 튜플을 사용하는 이유
- 추가, 삭제, 변경을 할 수 없다
    - 여러명이 코딩을 할 때 변경하면 안되는 변수를 못 건드리게 하고
    - 값이 변경되면 안되는 중요한 데이터를 넣어서 보안적으로 관리하기 좋게함
- list가 tuple보다 무겁다
    - 단지 list보다 기능이 적기 때문이 아니라  
    파이썬은 c기반이고 동적배열을 지원하는데 list같은 경우 생성한 저장소를 다 사용할 경우 추가할 때 추가 할당을 하는 데  
    이때 4, 8, 16... 처럼 더블링을 하는 것 때문에 더 무겁다
- list는 가변이고 tuple은 불변이기 때문에 값이 고정되어야 하거나 크기가 작을 경우 튜플을 권장한다

```python
tuple1 = (1,"이",3.0)
tuple2 = 1,"이",3.0   #튜플만 가능
tuple3 = (1,)        #1개의 값만 들어있는 튜플

print(type(tuple1), type(tuple2))
print(dir(tuple1))    #튜플이 가진 메소드를 볼 수 있는데 리스트보다 적음

str1="abcd"
print(list(str1), tuple(str1))      #문자열의 문자들이 하나씩 요소가 됨
print("o".join(list(str1)))         #join함수는 !앞!의 "(구분)"기호를 넣어서 !뒤!의요소들을 합침

tuple1[2]
#tuple1[2] = 3      #인덱싱은 되지만 교체는 X
```

### 딕셔너리dict
- list, tuple, str, range같은 데이터 시퀸스의 요소들은 데이터의 연관성이 없는데 dict는 값끼리의 연관성이 있다

```python
#딕셔너리 생성
dict1={}
dict2=dict()
dict3=dict(힘=30, 지능=20, 체력=40)                 #기호주의!!
dict4=dict(zip(["힘","지능","체력"],[30,20,40]))    #zip()병렬처리를 도와주는 함수
hero={"힘":30,"지능":20, "체력":40}                 #{키:값}으로 이루어졌다

#hero[1]
hero["힘"]      #키로 값을 찾기때문에 값은 중복을 허락하지만 키는 허락X -> 키가 중복된다면 뒤의 것으로 적용

#딕셔너리 값 확인
print("힘" in hero)
print(hero.keys())
print(hero.values())
print(hero.items())
print(type(hero.items()))
listHero=list(hero.items())    #dict_items를 리스트화 하면 인덱싱이 가능해 값을 뽑기 편해진다
print(listHero[0][1])

#딕셔너리 값 추가, 변경, 삭제
hero.update({"민첩":90, "손재주":44})
hero["지능"]=100
del hero["민첩"]
print(hero)
```

### set
- 수학의 집합
- 딕셔너리와 같이 순서가 없다, 중괄호를 사용한다
- 값의 중복을 허락하지 않는다

```python
set1={}                   #type 'dict'나옴
set2=set()
채소={"당근","배추","대파"}
type(채소)
print(채소)                 #출력할 때마다 요소가 다르게 나오는 것이 특징 (순서가 바뀜)

print(list("apple"))
print(set("apple"))        #요소의 중복이 없음

#set의 값 추가, 삭제, 변경
채소.add('토마토')
채소.update('깻잎')          #깻잎이 따로따로 추가됨
print(채소)

채소.remove("잎")
#채소.remove("잎")          #romove는 중복될 때 에러나지만 discard는 중복되도 에러가 나지 않음
채소.discard("깻")
채소.discard("깻")
print(채소.pop())          #set은 임의의 요소를, 나머지는 끝은 요소를 반환하면서 삭제하는 함수 'pop()'
print(채소)
```

```python
set1={'a','b','c','d'}
set2={'b','c','d','e'}

#합집합 : 둘 다 가능하다 set으로 반환함
set1 | set2
set.union(set1, set2)

#교집합
set1 & set2
set.intersection(set1, set2)

#차집합
set1 - set2
set.difference(set1, set2)
```

```python
# 문자열 startswith : 시작하는 값 확인
print("python".startswith("py"))

# find
print("python hello".find("world"))     # 값이 없으면 -1, 있으면 7(시작 인덱스를 반환)
# in : 왼쪽 값을 오른쪽 값에서 검색
print("a" in "aaple")

# islower / isupper : 소문자로 이루어졌는지 대문자로 이루어졌는지 판단
print("hello".islower())
print("hello".isupper())

# upper / lower : 대문자/소문자 변환
print("hello".upper())
print("HELLO".lower())
```
